```python
c = input("Nhập chữ cái hoa (A-Z): ")
ma_ascii_thuong = ord(c.lower())
if c == 'A':
    print('z')  
else:
    print(chr(ma_ascii_thuong - 1))
```

Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
import math

a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
c = float(input("Nhập c: "))

if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"{s:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

Bài 8: Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh
Tính thêm tiền VAT 8%. Chỉ lấy đến đơn vị đồng

```python
ten = input("Ten chu ho: ")
cu = int(input("Chi so thang truoc: "))
moi = int(input("Chi so thang nay: "))

sodu = moi - cu
tien_dien = 0

if sodu <= 50:
    tien_dien = sodu * 1984
elif sodu <= 100:
    tien_dien = 50 * 1984 + (sodu - 50) * 2050
elif sodu <= 200:
    tien_dien = 50 * 1984 + 50 * 2050 + (sodu - 100) * 2380
elif sodu <= 300:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + (sodu - 200) * 2998
elif sodu <= 400:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (sodu - 300) * 3350
else:
    tien_dien = 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (sodu - 400) * 3460

# Cộng VAT 8% và làm tròn về số nguyên (đơn vị đồng)
tong_tien = round(tien_dien * 1.08)

print(f"Ho va ten: {ten}")
print(f"Tien phai tra la: {tong_tien}")
```

Kiểm tra tuổi đủ điều kiện bầu cử: Viết chương trình nhập năm sinh của một
người, kiểm tra xem người đó có đủ 18 tuổi để đi bầu cử hay không. 


```python
from datetime import datetime
nam_sinh = int(input("Nhập năm sinh: "))
nam_hien_tai = datetime.now().year 
tuoi = nam_hien_tai - nam_sinh
if tuoi >= 18:
    print("Đủ điều kiện bầu cử")
else:
    print("Chưa đủ điều kiện bầu cử")
```

Kiểm tra năm nhuận: Viết chương trình nhập năm, kiểm tra xem năm đó có phải
năm nhuận hay không. (Năm nhuận nếu chia hết cho 400 hoặc chia hết cho 4
nhưng không chia hết cho 100)

```python
nam = int(input("Nhập năm cần kiểm tra: "))

if (nam % 400 == 0) or (nam % 4 == 0 and nam % 100 != 0):
    print(f"Năm {nam} là năm nhuận")
else:
    print(f"Năm {nam} không phải năm nhuận")
```

Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
n = int(input("Nhập một số nguyên từ 0 đến 9: "))

match n:
    case 0: print("Không")
    case 1: print("Một")
    case 2: print("Hai")
    case 3: print("Ba")
    case 4: print("Bốn")
    case 5: print("Năm")
    case 6: print("Sáu")
    case 7: print("Bảy")
    case 8: print("Tám")
    case 9: print("Chín")
    case _: print("Số nhập vào không hợp lệ")
```
