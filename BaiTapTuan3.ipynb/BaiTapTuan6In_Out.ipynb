Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới
trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có
trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của

```python
import ast
dict1 = ast.literal_eval(input())
dict2 = ast.literal_eval(input())

result = dict1.copy()
for key, value in dict2.items():
    result[key] = result.get(key, 0) + value
sorted_result = {k: result[k] for k in sorted(result.keys())}
print(sorted_result)
```

W6A10
Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra
list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple
có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của
số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên
trong tuple

```python
arr = list(map(int, input().split()))
k = int(input())

pairs = []
n = len(arr)

for i in range(n):
    for j in range(i + 1, n):
        if arr[i] + arr[j] == k:
            pairs.append((arr[i], arr[j]))
pairs.sort(key=lambda x: x[0])
print(pairs)
```

W6A17
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
đường chéo phụ của mảng

```python
def w6a17():
    n = int(input())
    matrix = []
    for _ in range(n):
        matrix.append(list(map(int, input().split())))
    main_diagonal = [matrix[i][i] for i in range(n)]
    side_diagonal = [matrix[i][n - 1 - i] for i in range(n)]
    
    print(*(main_diagonal))
    print(*(side_diagonal))
```

W6A18
Nhâ ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột
k (hàng k) 


```python
def w6a18():
    n, m, k = map(int, input().split())
    
    matrix = []
    for _ in range(n):
        matrix.append(list(map(int, input().split())))
    sum_row = sum(matrix[k]) if 0 <= k < n else 0
    sum_col = sum(matrix[i][k] for i in range(n)) if 0 <= k < m else 0
    
    print(f"Tổng hàng {k}: {sum_row}")
    print(f"Tổng cột {k}: {sum_col}")
```

W6A24
Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def w6a24_move_zeros(arr, to_end=True):
    zeros = [num for num in arr if num == 0]
    non_zeros = [num for num in arr if num != 0]
    
    if to_end:
        return non_zeros + zeros
    else:
        return zeros + non_zeros
```

W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
phần tử đầu và cuối

```python
def w6a26(lst):
    total_sum = 0
    for i in range(1, len(lst) - 1):
        if lst[i] > lst[i - 1] and lst[i] > lst[i + 1]:
            total_sum += lst[i]
            
    return total_sum
```

W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
def w6a28(arr):
    if len(arr) < 2:
        return "vô danh"
        
    is_increasing = True
    is_decreasing = True
    
    for i in range(len(arr) - 1):
        if arr[i] >= arr[i + 1]:
            is_increasing = False
        if arr[i] <= arr[i + 1]:
            is_decreasing = False
            
    if is_increasing:
        return "dãy tăng thực sự"
    elif is_decreasing:
        return "dãy giảm thực sự"
    else:
        return "vô danh"
```
