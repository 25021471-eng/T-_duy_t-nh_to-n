Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h = int(input("\nBài 4 - Nhập độ cao h: "))
for i in range(1, h + 1):
    if i == 1:
        print(' ' * (h - i) + '*')
    elif i == h:
        print('*' * (2 * h - 1))
    else:
        print(' ' * (h - i) + '*' + ' ' * (2 * i - 3) + '*')
```

Bài 9: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số chính phương hay
không? Số chính phương là số có căn bậc hai là một số nguyên VD: 1, 4, 9, 16, 25…. 


```python
a = int(input("\nBài 9 - Nhập a: "))
is_square = False
i = 1
while i * i <= a:
    if i * i == a:
        is_square = True
        break
    i += 1
if is_square:
    print(f"{a} là số chính phương.")
else:
    print(f"{a} không phải số chính phương.")
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 


```python
a = int(input("\nBài 10 - Nhập a: "))
S_uoc = 0
for i in range(1, a):
    if a % i == 0:
        S_uoc += i
if S_uoc == a and a > 0:
    print(f"{a} là số hoàn hảo.")
else:
    print(f"{a} không phải số hoàn hảo.")
```

Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k)
lớn nhất nhưng nhỏ hơn n 


```python
n = int(input("\nBài 12 - Nhập n: "))
k = 0
S_k = 0
while S_k + (k + 1) < n:
    k += 1
    S_k += k
print(f"Giá trị k lớn nhất là: {k} (Với S(k) = {S_k})")
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho
1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 

```python
A = float(input("\nBài 13 - Nhập A: "))
n = 1
S_A = 1.0
while S_A <= A:
    n += 1
    S_A += 1 / n
print(f"Giá trị n nhỏ nhất là: {n}")
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng 

```python
print("\nBài 14:")
first = True
max_val = 0
min_val = 0
while True:
    val = int(input("Nhập một số (nhập -1 để dừng): "))
    if val == -1:
        break
    if first:
        max_val = val
        min_val = val
        first = False
    else:
        if val > max_val:
            max_val = val
        if val < min_val:
            min_val = val

if not first:
    print(f"Số lớn nhất: {max_val}, Số nhỏ nhất: {min_val}")
else:
    print("Bạn chưa nhập số nào hợp lệ.")
```

Bài 15: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số. Làm bằng nhiều
cách nhất có thể, ko dùng mảng.

```python
n = int(input("\nBài 15 - Nhập n: "))
count_digits = 0
temp = n
if temp == 0:
    count_digits = 1
while temp > 0:
    count_digits += 1
    temp //= 10
print(f"Số lượng chữ số của {n} là: {count_digits}")
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu
chữ số lẻ 


```python
n = int(input("\nBài 16 - Nhập n: "))
even_count = 0
odd_count = 0
temp = n
if temp == 0:
    even_count = 1
while temp > 0:
    digit = temp % 10
    if digit % 2 == 0:
        even_count += 1
    else:
        odd_count += 1
    temp //= 10
print(f"Chữ số chẵn: {even_count}, Chữ số lẻ: {odd_count}")
```

Bài 17: Nhập vào số nguyên dương n, tính tổng các chữ số của n

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay
không 


```python
n = int(input("\nBài 18 - Nhập n: "))
temp = n
is_pow3 = True
if temp <= 0:
    is_pow3 = False
else:
    while temp > 1:
        if temp % 3 != 0:
            is_pow3 = False
            break
        temp //= 3
if is_pow3:
    print(f"{n} là số có dạng 3^k.")
else:
    print(f"{n} không phải là số có dạng 3^k.")
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input("\nBài 19 - Nhập A: "))
f1 = 1
f2 = 1
max_fibo = 1
while f2 <= A:
    max_fibo = f2
    next_fibo = f1 + f2
    f1 = f2
    f2 = next_fibo
print(f"Số Fibonacci lớn nhất không vượt quá {A} là: {max_fibo}")
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.


```python
n = int(input("\nBài 20 - Nhập n: "))
prime_digit_count = 0
temp = n
while temp > 0:
    digit = temp % 10
    if digit == 2 or digit == 3 or digit == 5 or digit == 7:
        prime_digit_count += 1
    temp //= 10
print(f"Số lượng chữ số là số nguyên tố của {n} là: {prime_digit_count}")
```
