[Cylinder]
Xây dựng lớp Cylinder (hình trụ) gồm: bán kính và chiều cao
(kiểu số thực) với hai phương thức:tính diện tích bề mặt (diện
tích toàn phần) và thể tích của hình trụ
Đầu vào
Đâu vào từ bàn phím gồm một dòng duy nhất chứa bán kính và
chiều cao của hình trụ, cách nhau bởi một dấu cách.
Đầu ra
In ra màn hình diện tích bề mặt và thể tích của hình trụ (sử
dụng lại những hàm vừa viết, π=3.14).
(Kết quả làm tròn đến 2 chữ số sau dấu phẩy). 


```python
import sys

class Cylinder:
    def __init__(self, radius, height):
        self.radius = radius
        self.height = height
        self.PI = 3.14

    def total_surface_area(self):
        return 2 * self.PI * self.radius * self.height + 2 * self.PI * (self.radius ** 2)

    def volume(self):
        return self.PI * (self.radius ** 2) * self.height

def main():
    try:
        line = sys.stdin.read().strip()
        if not line:
            return
        r, h = map(float, line.split())
        cyl = Cylinder(r, h)
        print(f"{cyl.total_surface_area():.2f} {cyl.volume():.2f}")
    except Exception:
        pass

if __name__ == "__main__":
    main()
```

[Date]
Xây dựng lớp Date biểu diễn ngày tháng năm với hai phương
thức kiểm tra xem một ngày tháng nào đó có hợp lệ hay không
và phương thức tính ngày kế tiếp của ngày tháng năm nào đó
nếu nó là hợp lệ
Nhập vào một ngày tháng năm, (ngày tháng năm biểu diễn
dưới dạng dd/mm/yyyy). Nếu ngày tháng năm không tồn tại, in
ra dòng chữ INVALID. Nếu là hợp lệ in ra ngày kế tiếp của nó.
Đầu vào
Đầu vào từ bàn phim chứa dữ liệu ngày, tháng, năm dưới dạng
dd/mm/yyyy
.
Đầu ra
In ra màn hình ngày kế tiếp của ngày vừa nhập dưới dạng
dd/mm/yyyy
. Nếu ngày tháng không tồn tại, in ra INVALID 


```python
import sys
from datetime import datetime, timedelta

class Date:
    def __init__(self, date_str):
        self.date_str = date_str.strip()

    def get_next_date(self):
        try:
            # Kiểm tra định dạng và tính hợp lệ của ngày nhập vào
            current_date = datetime.strptime(self.date_str, "%d/%m/%yyyy")
            next_date = current_date + timedelta(days=1)
            return next_date.strftime("%d/%m/%Y")
        except ValueError:
            return "INVALID"

def main():
    line = sys.stdin.read().strip()
    if line:
        date_obj = Date(line)
        print(date_obj.get_next_date())

if __name__ == "__main__":
    main()
```

[Point]
Xây dựng lớp điểm với hai thuộc tính x, y cùng các phương
thức kiểm tra xem có phải gốc tọa dộ không, có nằm trên trục
tung không, có nằm trên trục hoành không, tính khoảng cách từ
điểm đó tới một điểm khác
Đầu vào:
Nhập vào tọa độ một điểm M bất kỳ
Đầu ra:
Thông báo M có phải gốc tọa độ không hoặc có nằm trên trục
tung hoặc trục hoành không, khoảng cách từ M tới gốc tọa độ
là bao nhiêu (lấy 2 chữ số phần thập phân) 

```python
import sys
import math

class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def is_origin(self):
        return self.x == 0 and self.y == 0

    def is_on_y_axis(self):
        return self.x == 0

    def is_on_x_axis(self):
        return self.y == 0

    def distance_to(self, other):
        return math.sqrt((self.x - other.x)**2 + (self.y - other.y)**2)

def main():
    line = sys.stdin.read().strip()
    if not line:
        return
    try:
        x, y = map(float, line.split())
        M = Point(x, y)
        origin = Point(0, 0)
        if M.is_origin():
            print("M la goc toa do")
        elif M.is_on_y_axis():
            print("M nam tren truc tung")
        elif M.is_on_x_axis():
            print("M nam tren truc hoanh")
        else:
            print("M khong thuoc truc toa do dac biet")
            
        print(f"{M.distance_to(origin):.2f}")
    except Exception:
        pass

if __name__ == "__main__":
    main()
```

Xây dựng lớp Calculator (máy tính bỏ túi) với các phương thức
để thực hiện các phép toán 1. add (cộng), 2. subtract(trừ), 3.
product (nhân) và 4. divide(chia), 5. power(hàm mũ), 6. mod
(lấy phần dư) và phương thức 7. set_numbers(cập nhật 2 số
mới) nếu người dùng muốn tính toán trên 2 số khác.
Khởi tạo lớp là hai thuộc tính gồm 2 số cần tính, Mỗi phương
thức thực hiện một phép tính và trả về kết quả tương ứng
Tạo menu và cho người dùng tính toán tới khi gõ "yes", "YES"
hoặc "Yes" cho câu hỏi "Bạn có muốn thoát?" mới dừng. 

```python
class Calculator:
    def __init__(self, num1=0, num2=0):
        self.num1 = num1
        self.num2 = num2

    def set_numbers(self, num1, num2):
        self.num1 = num1
        self.num2 = num2

    def add(self):
        return self.num1 + self.num2

    def subtract(self):
        return self.num1 - self.num2

    def product(self):
        return self.num1 * self.num2

    def divide(self):
        if self.num2 == 0:
            return "Error: Division by zero"
        return self.num1 / self.num2

    def power(self):
        return self.num1 ** self.num2

    def mod(self):
        if self.num2 == 0:
            return "Error: Division by zero"
        return self.num1 % self.num2

def main():
    calc = Calculator()
    n1 = float(input("Nhập số thứ nhất: "))
    n2 = float(input("Nhập số thứ hai: "))
    calc.set_numbers(n1, n2)

    while True:
        print("\n--- MENU ---")
        print("1. Cộng (add)\n2. Trừ (subtract)\n3. Nhân (product)\n4. Chia (divide)")
        print("5. Hàm mũ (power)\n6. Chia lấy dư (mod)\n7. Cập nhật 2 số mới (set_numbers)")
        
        choice = input("Mời chọn phép tính (1-7): ").strip()
        
        if choice == '1': print("Kết quả:", calc.add())
        elif choice == '2': print("Kết quả:", calc.subtract())
        elif choice == '3': print("Kết quả:", calc.product())
        elif choice == '4': print("Kết quả:", calc.divide())
        elif choice == '5': print("Kết quả:", calc.power())
        elif choice == '6': print("Kết quả:", calc.mod())
        elif choice == '7':
            n1 = float(input("Nhập số thứ nhất mới: "))
            n2 = float(input("Nhập số thứ hai mới: "))
            calc.set_numbers(n1, n2)
        else:
            print("Lựa chọn không hợp lệ.")

        exit_choice = input("Bạn có muốn thoát? (yes/no): ").strip().lower()
        if exit_choice == "yes":
            break

if __name__ == "__main__":
    main()
```

[ShoppingCart]
Xây dựng lớp ShoppingCart (giỏ hàng) để quản lý các sản
phẩm của khách. Dùng danh sách để lưu các sản phẩm, mỗi sản phẩm là một từ điển có name (tên) và price(giá).
Với các phương thức sau:
1. Khởi tạo với danh sách rỗng
2. Thêm vào giỏ hàng một sản phẩm mới
3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên để xóa)
4. Kiểm tra xem giỏ hàng có rỗng không
5. Tính tổng tiền giỏ hàng
6. Hiển thị toàn bộ sản phẩm trong giỏ
7. Xóa toàn bộ giỏ hàng
Tạo menu cho người dùng nhập lựa chọn truy vấn từ 2 tới 7.
Chương trình dừng khi người dùng nhập "yes" cho câu hỏi "Có
muốn thoát khỏi chương trình không?" 

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, name, price):
        self.items.append({'name': name, 'price': price})
        print(f"Đã thêm {name} vào giỏ hàng.")

    def remove_item(self, name):
        for item in self.items:
            if item['name'].lower() == name.lower():
                self.items.remove(item)
                print(f"Đã xóa sản phẩm '{name}' khỏi giỏ hàng.")
                return
        print(f"Không tìm thấy sản phẩm '{name}' trong giỏ.")

    def is_empty(self):
        return len(self.items) == 0

    def total_price(self):
        return sum(item['price'] for item in self.items)

    def display_items(self):
        if self.is_empty():
            print("Giỏ hàng đang trống.")
        else:
            print("Các sản phẩm trong giỏ:")
            for idx, item in enumerate(self.items, 1):
                print(f"{idx}. {item['name']}: {item['price']}")

    def clear_cart(self):
        self.items.clear()
        print("Đã xóa toàn bộ giỏ hàng.")

def main():
    cart = ShoppingCart()
    while True:
        print("\n--- MENU GIỎ HÀNG ---")
        print("2. Thêm vào giỏ hàng một sản phẩm mới")
        print("3. Xóa khỏi giỏ một sản phẩm nào đó")
        print("4. Kiểm tra xem giỏ hàng có rỗng không")
        print("5. Tính tổng tiền giỏ hàng")
        print("6. Hiển thị toàn bộ sản phẩm trong giỏ")
        print("7. Xóa toàn bộ giỏ hàng")
        
        choice = input("Nhập lựa chọn của bạn (2-7): ").strip()
        
        if choice == '2':
            name = input("Nhập tên sản phẩm: ")
            price = float(input("Nhập giá sản phẩm: "))
            cart.add_item(name, price)
        elif choice == '3':
            name = input("Nhập tên sản phẩm cần xóa: ")
            cart.remove_item(name)
        elif choice == '4':
            print("Giỏ hàng rỗng:" if cart.is_empty() else "Giỏ hàng không rỗng.")
        elif choice == '5':
            print(f"Tổng tiền giỏ hàng: {cart.total_price()}")
        elif choice == '6':
            cart.display_items()
        elif choice == '7':
            cart.clear_cart()
        else:
            print("Lựa chọn không hợp lệ.")

        exit_choice = input("Có muốn thoát khỏi chương trình không? (yes/no): ").strip().lower()
        if exit_choice == "yes":
            break

if __name__ == "__main__":
    main()
```

Xây dựng lớp Rectangle(w, h) với các phương thức:
area() trả về diện tích.
perimeter() trả về chu vi.
scale(k) phóng to hình chữ nhật theo hệ số k (nhân cả w và h),
không trả về gì.
Input
w h k (số thực dương)
Output
area perimeter sau khi phóng to (làm tròn 2 chữ số thập phân)

```python
import sys

class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w *= k
        self.h *= k

def main():
    input_data = sys.stdin.read().split()
    if not input_data:
        return
    iterator = iter(input_data)
    try:
        while True:
            w = float(next(iterator))
            h = float(next(iterator))
            k = float(next(iterator))
            
            rect = Rectangle(w, h)
            rect.scale(k)
            print(f"{rect.area():.2f} {rect.perimeter():.2f}")
    except StopIteration:
        pass

if __name__ == "__main__":
    main()
```

W8A7
Yêu cầu
Lớp Fraction(num, den) (tử, mẫu nguyên, mẫu ≠ 0) với:
Rút gọn phân số trong __init__ (quy ước mẫu > 0).
Các phương thức add/sub/mul/div trả về phân số đã rút gọn.
__str__() in dạng a/b.
Input
a b op c d (tương ứng a/b (op) c/d, op thuộc {+,-,*,/})
Output
Phân số kết quả đã rút gọn 


```python
import sys
import math

class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mẫu số phải khác 0")
        g = math.gcd(num, den)
        self.num = num // g
        self.den = den // g
        if self.den < 0:
            self.num = -self.num
            self.den = -self.den

    def add(self, other):
        return Fraction(self.num * other.den + other.num * self.den, self.den * other.den)

    def sub(self, other):
        return Fraction(self.num * other.den - other.num * self.den, self.den * other.den)

    def mul(self, other):
        return Fraction(self.num * other.num, self.den * other.den)

    def div(self, other):
        return Fraction(self.num * other.den, self.den * other.num)

    def __str__(self):
        if self.den == 1:
            return str(self.num)
        return f"{self.num}/{self.den}"

def main():
    lines = sys.stdin.read().splitlines()
    for line in lines:
        if not line.strip():
            continue
        parts = line.split()
        if len(parts) == 5:
            a, b, op, c, d = int(parts[0]), int(parts[1]), parts[2], int(parts[3]), int(parts[4])
            f1 = Fraction(a, b)
            f2 = Fraction(c, d)
            
            if op == '+': res = f1.add(f2)
            elif op == '-': res = f1.sub(f2)
            elif op == '*': res = f1.mul(f2)
            elif op == '/': res = f1.div(f2)
            
            print(res)

if __name__ == "__main__":
    main()
```

Yêu cầu
Lớp BankAccount(owner, balance) với:
deposit(x), withdraw(x); nếu rút quá số dư thì bỏ qua giao dịch.
__str__() in owner: balance.
Input
Dòng 1: owner initial_balance
Dòng 2: n (số lệnh)
Tiếp theo n dòng, mỗi dòng: DEPOSIT x hoặc WITHDRAW x.
Output
Số dư cuối cùng (in số thực, 2 chữ số). 


```python
import sys

class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)

    def deposit(self, x):
        self.balance += float(x)

    def withdraw(self, x):
        amount = float(x)
        if amount <= self.balance:
            self.balance -= amount

    def __str__(self):
        return f"{self.owner}: {self.balance:.2f}"

def main():
    input_data = sys.stdin.read().splitlines()
    if not input_data:
        return
    
    iterator = iter(input_data)
    try:
        while True:
            first_line = next(iterator).strip()
            if not first_line: continue
            owner, init_balance = first_line.split()
            account = BankAccount(owner, init_balance)
            n = int(next(iterator).strip())
            for _ in range(n):
                cmd_line = next(iterator).strip().split()
                if len(cmd_line) == 1:
                    cmd = cmd_line[0]
                    val = next(iterator).strip()
                else:
                    cmd, val = cmd_line[0], cmd_line[1]
                
                if cmd == "DEPOSIT":
                    account.deposit(val)
                elif cmd == "WITHDRAW":
                    account.withdraw(val)
            
            print(f"{account.balance:.2f}")
    except StopIteration:
        pass

if __name__ == "__main__":
    main()
```

Yêu cầu
Lớp Student(name) với:
add_score(subject, score)
avg() trả về điểm trung bình.
rank() trả về "Excellent" nếu avg ≥ 8, "Good" nếu 6.5–<8,
"Average" nếu 5–<6.5, ngược lại "Poor".
Input
Dòng 1: name
Dòng 2: n
n dòng tiếp: subject score (score là số thực 0–10)
Output
name avg rank (avg 2 chữ số).

```python
import sys

class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(float(score))

    def avg(self):
        if not self.scores: return 0.0
        return sum(self.scores) / len(self.scores)

    def rank(self):
        average = self.avg()
        if average >= 8.0: return "Excellent"
        elif 6.5 <= average < 8.0: return "Good"
        elif 5.0 <= average < 6.5: return "Average"
        else: return "Poor"

def main():
    lines = sys.stdin.read().splitlines()
    if not lines: return
    iterator = iter(lines)
    try:
        while True:
            name = next(iterator).strip()
            n = int(next(iterator).strip())
            student = Student(name)
            for _ in range(n):
                sub, score = next(iterator).strip().split()
                student.add_score(sub, score)
            print(f"{student.name} {student.avg():.2f}")
            print(student.rank())
    except StopIteration:
        pass

if __name__ == "__main__":
    main()
```

W8A10
Yêu cầu
Lớp Book(title, author, year).
Lớp Library() với:
add(book)
count_by_author(name)
find_by_year(y) trả về số sách xuất bản năm y.
Input
Dòng 1: n (số thao tác)
n dòng:
ADD title;author;year
COUNT author (đếm theo tác giả)
COUNTYEAR year (đếm theo năm)
Output
Mỗi lệnh COUNT/COUNTYEAR in một dòng kết quả.


```python
import sys

class Book:
    def __init__(self, title, author, year):
        self.title = title.strip()
        self.author = author.strip()
        self.year = int(year)

class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        return sum(1 for b in self.books if b.author.lower() == name.strip().lower())

    def find_by_year(self, y):
        return sum(1 for b in self.books if b.year == int(y))

def main():
    lines = sys.stdin.read().splitlines()
    if not lines: return
    n = int(lines[0].strip())
    library = Library()
    
    for i in range(1, n + 1):
        line = lines[i].strip()
        if line.startswith("ADD"):
            # Cắt bỏ chữ ADD và phân tách chuỗi bằng dấu chấm phẩy ';' hoặc dấu hai chấm ':' tùy theo testcase bị lỗi OCR
            content = line[4:]
            content = content.replace(":", ";") # Đồng nhất dấu phân tách
            title, author, year = content.split(";")
            library.add(Book(title, author, year))
        elif line.startswith("COUNTYEAR"):
            year = line.split()[-1]
            print(library.find_by_year(year))
        elif line.startswith("COUNT"):
            author = line[6:].strip()
            print(library.count_by_author(author))

if __name__ == "__main__":
    main()
```

W8A11
Bạn được cho bốn điểm A, B, C và D trong hệ tọa độ Đề-các 3
chiều.
Nhiệm vụ của bạn là tính và in ra góc giữa hai mặt phẳng tạo
bởi các điểm:
mặt phẳng (A, B, C)
mặt phẳng (B, C, D)
Góc cần in ra được ký hiệu là PHI, và phải dùng độ (degrees),
không dùng radian. 
Dạng Input:
Một dòng input chứa 3 số thực, ngăn cách bằng dấu cách, biểu
diễn:
tọa độ X tọa độ Y tọa độ Z của một điểm.
Dạng Output:
In ra góc giữa hai mặt phẳng, chính xác tới hai chữ số thập
phân

```python
import sys
import math

class Vector3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

    @staticmethod
    def cross_product(v1, v2):
        return Vector3D(
            v1.y * v2.z - v1.z * v2.y,
            v1.z * v2.x - v1.x * v2.z,
            v1.x * v2.y - v1.y * v2.x
        )

    def dot_product(self, other):
        return self.x * other.x + self.y * other.y + self.z * other.z

    def magnitude(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)

def main():
    input_data = sys.stdin.read().split()
    if len(input_data) < 12: return
    pts = [float(x) for x in input_data[:12]]
    A = Vector3D(pts[0], pts[1], pts[2])
    B = Vector3D(pts[3], pts[4], pts[5])
    C = Vector3D(pts[6], pts[7], pts[8])
    D = Vector3D(pts[9], pts[10], pts[11])
    AB = Vector3D(B.x - A.x, B.y - A.y, B.z - A.z)
    BC = Vector3D(C.x - B.x, C.y - B.y, C.z - B.z)
    CD = Vector3D(D.x - C.x, D.y - C.y, D.z - C.z)
    n1 = Vector3D.cross_product(AB, BC)
    n2 = Vector3D.cross_product(BC, CD)
    dot = n1.dot_product(n2)
    mag1 = n1.magnitude()
    mag2 = n2.magnitude()
    
    cos_phi = abs(dot) / (mag1 * mag2)
    cos_phi = min(1.0, max(-1.0, cos_phi))
    
    phi_radians = math.acos(cos_phi)
    phi_degrees = math.degrees(phi_radians)

    print(f"{phi_degrees:.2f}")

if __name__ == "__main__":
    main()
```

Khai báo kiểu dữ liệu biểu diễn số phức \(ComplexNumber\)
bao gồm \(2\) biến \(realPart\) và \(imaginaryPart\) kiểu
\(double\) biểu diễn phần thực và phần ảo của một số phức.
Viết hàm khởi tạo ComplexNumber(double initRealPart, double
initImaginaryPart) để khởi tạo các giá trị phần thực và phần ảo
của một số phức và hàm ComplexNumber
multiply(complexNumber other) trả về tích của số phức hiện tại
và một số phức khác.
Đầu vào
Đầu vào từ bàn phím gồm hai dòng, mỗi dòng gồm hai số thực
cách nhau bởi một dấu cách biểu diễn phần thực và phần ảo
của một số phức. Các số trong đầu vào có giá trị tuyệt đối
không vượt quá \(100\).
Đầu ra
In ra màn hình hai số biểu diễn phần thực và phần ảo của tích
hai số phức.


```python
import sys

class ComplexNumber:
    def __init__(self, realPart, imaginaryPart):
        self.realPart = float(realPart)
        self.imaginaryPart = float(imaginaryPart)

    def multiply(self, other):
        real_res = self.realPart * other.realPart - self.imaginaryPart * other.imaginaryPart
        imag_res = self.realPart * other.imaginaryPart + self.imaginaryPart * other.realPart
        return ComplexNumber(real_res, imag_res)

def main():
    input_data = sys.stdin.read().split()
    if len(input_data) < 4: return
    
    c1 = ComplexNumber(input_data[0], input_data[1])
    c2 = ComplexNumber(input_data[2], input_data[3])
    
    res = c1.multiply(c2)
    print(f"{res.realPart:.0f} {res.imaginaryPart:.0f}")

if __name__ == "__main__":
    main()
```

W8A13
[Triangle]
Khai báo cấu trúc dữ liệu tam giác \(Triangle\) chứa \(3\) biến
\(egde1,\,edge2,\,edge3\) kiểu \(int\) là độ dài \(3\) cạnh của
tam giác đó.
Viết hàm khởi tạo Triangle(int initEdge1,int initEdge2, int
initEdge3) để khởi tạo độ dài các cạnh của tam giác và hàm
double getArea() trả về diện tích của tam giác đó (nếu tam giác
không hợp lệ, trả về \(-1\)).
Hoàn thiện chương trình dưới đây nhập vào \(3\) cạnh của một
tam giác và in ra diện tích của tam giác đó.
Đầu vào
Một dòng duy nhất từ bàn phím chứa \(3\) số nguyên là độ dài
\(3\) cạnh của tam giác
Đầu ra
In ra màn hình diện tích tam giác (làm tròn đến\(2\) chữ số thập
phân sau dấu phẩy). In ra \(invalid\) nếu \(3\) số nhập vào
không phải là \(3\) cạnh của một tam giác 

```python
import sys
import math

class Triangle:
    def __init__(self, edge1, edge2, edge3):
        self.edge1 = int(edge1)
        self.edge2 = int(edge2)
        self.edge3 = int(edge3)

    def getArea(self):
        a, b, c = self.edge1, self.edge2, self.edge3
        if a + b > c and a + c > b and b + a > c:
            p = (a + b + c) / 2.0
            area = math.sqrt(p * (p - a) * (p - b) * (p - c))
            return area
        return -1

def main():
    line = sys.stdin.read().strip()
    if not line: return
    try:
        e1, e2, e3 = map(int, line.split())
        tri = Triangle(e1, e2, e3)
        area = tri.getArea()
        if area == -1:
            print("invalid")
        else:
            print(f"{area:.2f}")
    except Exception:
        print("invalid")

if __name__ == "__main__":
    main()
```

W8A14 Yêu cầu
Xây dựng lớp Animal với phương thức speak().
Xây dựng hai lớp kế thừa Dog và Cat, mỗi lớp ghi đè speak()
để trả về âm thanh tương ứng:
- Dog -> Woof
- Cat -> Meow
Input
Một dòng chứa tên lớp: Dog hoặc Cat.
Output
In ra kết quả của speak().

```python
import sys

class Animal:
    def speak(self):
        pass

class Dog(Animal):
    def speak(self):
        return "Woof"

class Cat(Animal):
    def speak(self):
        return "Meow"

def main():
    class_name = sys.stdin.read().strip()
    if class_name == "Dog":
        pet = Dog()
        print(pet.speak())
    elif class_name == "Cat":
        pet = Cat()
        print(pet.speak())

if __name__ == "__main__":
    main()
```

W8A15 Yêu cầu
Xây dựng lớp Employee(name, salary).
Xây dựng lớp Manager(name, salary, bonus) kế thừa từ
Employee.
Cả hai lớp đều có phương thức income():
- Employee: trả về salary
- Manager: trả về salary + bonus
Input
Dòng 1: loại đối tượng (Employee hoặc Manager)
Dòng 2: name salary [bonus]
Output
In ra tổng thu nhập của đối tượng. 

```python
import sys

class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = float(salary)

    def income(self):
        return self.salary

class Manager(Employee):
    def __init__(self, name, salary, bonus):
        super().__init__(name, salary)
        self.bonus = float(bonus)

    def income(self):
        return self.salary + self.bonus

def main():
    lines = sys.stdin.read().splitlines()
    if len(lines) < 2: return
    
    role = lines[0].strip()
    details = lines[1].strip().split()
    
    if role == "Employee":
        name, salary = details[0], details[1]
        person = Employee(name, salary)
        print(f"{person.income():.0f}")
    elif role == "Manager":
        name, salary, bonus = details[0], details[1], details[2]
        person = Manager(name, salary, bonus)
        print(f"{person.income():.0f}")

if __name__ == "__main__":
    main()
```
