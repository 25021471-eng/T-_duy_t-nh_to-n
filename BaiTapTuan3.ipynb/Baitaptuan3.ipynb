Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ
NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử
lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý.


```python
n = int(input())
if 100 <= n <= 999:
    don_vi = n % 10
    chuc = (n // 10) % 10
    tram = n // 100
    so_dao = don_vi * 100 + chuc * 10 + tram
    print(so_dao)
else:
    pass
```

Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2
hay không. 


```python
n = int(input())

if n > 0 and (n & (n - 1)) == 0:
    print(f"{n} là lũy thừa của 2")
else:
    print(f"{n} không phải là lũy thừa của 2")
```


[Phân loại tam giác]
Nhập vào 3 số nguyên dương a,b,c là độ dài 3 cạnh của một tam giác. Kiểm
tra 3 số đã nhập có tạo thành một tam giác hợp lệ hay không. Nếu không in
ra "Không phải tam giác". Nếu có, phân loại tam giác đó là "Tam giác đều",
"Tam giác cân", hay "Tam giác thường"

```python
a = int(input())
b = int(input())
c = int(input())
if (a + b > c) and (a + c > b) and (b + c > a):
    if a == b == c:
        print("Tam giác đều")
    elif a == b or b == c or a == c:
        print("Tam giác cân")
    else:
        print("Tam giác thường")
else:
    print("Không phải tam giác")
```

[Giải phương trình bậc nhất]
Nhập vào hai số thực a và b của phương trình ax + b = 0. Tìm và in ra
nghiệm của phương trình, kết quả làm tròn đến số thập phân thứ 2. In ra "Vô
nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng.

```python
a = float(input())
b = float(input())

if a == 0:
    if b == 0:
        print("Vô số nghiệm")
    else:
        print("Vô nghiệm")
else:
    x = -b / a
    print(f"{x:.2f}")
```

[Rounding]
Nhập vào một số thực: in ra kết quả làm tròn lên, tròn xuống, làm tròn tới số
nguyên gần nhất (không được sử dụng hàm sẵn có của pyton). Các kết quả
cách nhau một dấu cách

```python
x = float(input())
phan_nguyen = int(x)
phan_thap_phan = x - phan_nguyen
tron_xuong = phan_nguyen
if phan_thap_phan > 0:
    tron_len = phan_nguyen + 1
else:
    tron_len = phan_nguyen
if phan_thap_phan >= 0.5:
    gan_nhat = phan_nguyen + 1
else:
    gan_nhat = phan_nguyen

print(f"{tron_len} {tron_xuong} {gan_nhat}")
```

Cấp số nhân
Cho 4 số a, b, c, d. Hãy kiểm tra xem 4 số này có thể theo thứ tự tạo thành 1
cấp số nhân với công bội nguyên theo đúng thứ tự a, b, c, d hay không?
Gợi ý : Tìm công bội (b / a) nếu b chia hết cho a, sau đó lấy b nhân công bội
và so sánh vs c, c nhân công bội và so sánh vs d

```python
a = int(input())
b = int(input())
c = int(input())
d = int(input())
if a != 0 and b % a == 0:
    q = b // a 
    if b * q == c and c * q == d:
        print("Cấp số nhân hợp lệ")
    else:
        print("Không phải cấp số nhân")
else:
    if a == 0 and b == 0 and c == 0 and d == 0:
        print("Cấp số nhân hợp lệ (Công bội tùy ý)")
    else:
        print("Không phải cấp số nhân")
```
