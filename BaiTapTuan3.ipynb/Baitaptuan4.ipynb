[SumOfNumBers]
Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để
tính tống các số nguyên từ 1 đến n. (n <= 1000)

```python
n = int(input("Nhập n (n <= 1000): "))
tong = 0
for i in range(1, n + 1):
    tong += i
print(f"Tổng các số từ 1 đến {n} là: {tong}")
```

[PrimeNumber]
VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi.
Kiểm tra số này có phải là số nguyên tố hay không? 

```python
import math
while True:
    n = int(input("Nhập một số nguyên dương: "))
    if n > 0:
        break
is_prime = True
if n < 2:
    is_prime = False
else:
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            is_prime = False
            break
if is_prime:
    print(f"{n} là số nguyên tố.")
else:
    print(f"{n} không phải là số nguyên tố.")
```

[Factorial]
Viết chương trình tính giai thừa của một số nguyên n cho trước (n!=1×2×…×n)
với 0 < n < 100. 


```python
n = int(input("Nhập n (0 < n < 100): "))
giai_thua = 1
for i in range(1, n + 1):
    giai_thua *= i
print(f"Giai thừa của {n} là: {giai_thua}")
```

.[CountingDigits]
Viết chương trình nhập vào một số nguyên n và in ra màn hình số chữ số (trừ dấu)
của số đó. Không dùng kiểu dữ liệu string.


```python
n = int(input("Nhập một số nguyên n: "))
n = abs(n)
count = 0

if n == 0:
    count = 1
else:
    while n > 0:
        count += 1
        n //= 10 

print(f"Số chữ số là: {count}")
```

"[SumPrimesInRange]
Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
import math

def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True
print("Nhập các bộ test (a b), nhấn Enter trống để dừng:")
while True:
    line = input()
    if not line:
        break
    a, b = map(int, line.split())
    
    tong_nguyen_to = 0
    for num in range(a, b + 1):
        if is_prime(num):
            tong_nguyen_to += num
            
    print(f"Tổng các số nguyên tố trong đoạn [{a}, {b}] là: {tong_nguyen_to}")
```

ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 


```python
n = int(input("Nhập n (n >= 2): "))
temp = n
max_prime_factor = -1
while temp % 2 == 0:
    max_prime_factor = 2
    temp //= 2
i = 3
while i * i <= temp:
    while temp % i == 0:
        max_prime_factor = i
        temp //= i
    i += 2
if temp > 1:
    max_prime_factor = temp

print(f"Ước số nguyên tố lớn nhất của {n} là: {max_prime_factor}")
```

[PerfectSquaresDistinctDigits]
Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu
cách). Nếu không có, in ra (no number).


```python
import math

n = int(input("Nhập số nguyên dương n: "))
results = []
max_base = int(math.sqrt(n))

for i in range(1, max_base + 1):
    chinh_phuong = i * i
    str_cp = str(chinh_phuong)
    if len(str_cp) == len(set(str_cp)):
        results.append(str_cp)
if results:
    print(" ".join(results))
else:
    print("(no number)")
```

"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất." 


```python
n = int(input("Nhập số nguyên dương n: "))

max_len = 0
best_x = 1

for x in range(1, n + 1):
    current = x
    length = 1
    
    while current != 1:
        if current % 2 == 0:
            current //= 2
        else:
            current = current * 3 + 1
        length += 1
        
    if length > max_len:
        max_len = length
        best_x = x

print(f"Số có dãy Collatz dài nhất: {best_x} {max_len}")
```

[EvenDivisors]
Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6.


```python
n = int(input("Nhập số nguyên dương n (< 10^6): "))
count = 0

for i in range(1, int(n**0.5) + 1):
    if n % i == 0:
        if i % 2 == 0:
            count += 1
        if (n // i) != i and (n // i) % 2 == 0:
            count += 1

print(f"Số lượng ước số chẵn của {n} là: {count}")
```

[ClosePairNumber]
Hai số nguyên dương tạo thành một cặp số thân thiết khi chúng tuân theo quy luật:
Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false" 


```python
def tinh_tong_uoc(num):
    tong = 0
    for i in range(1, num // 2 + 1):
        if num % i == 0:
            tong += i
    return tong

a = int(input("Nhập số thứ nhất: "))
b = int(input("Nhập số thứ hai: "))

if tinh_tong_uoc(a) == b and tinh_tong_uoc(b) == a:
    print("true")
else:
    print("false")
```

[GCD]
Cho hai số m và n là hai số nguyên dương nhập từ bàn phím. Viết chương trình
tìm ước chung lớn nhất của hai số m và n. 

```python
import math

m = int(input("Nhập số nguyên dương m: "))
n = int(input("Nhập số nguyên dương n: "))

ucln = math.gcd(m, n)

print(f"Ước chung lớn nhất của {m} và {n} là: {ucln}")
```

[ChickenAndDog]
"Vừa gà vừa chó,
Bó lại cho tròn,
Ba mươi sáu con,
Một trăm chân chẵn"
Từ bài toán dân gian trên, mở rộng thành chương trình nhận đầu vào là tổng số
con và tổng số chân của gà và chó. Nếu tìm được số phù hợp, in ra số lượng gà và
số lượng chó. Ngược lại, in ra "invalid".

```python
tong_con = int(input("Nhập tổng số con: "))
tong_chan = int(input("Nhập tổng số chân: "))
if (4 * tong_con - tong_chan) % 2 == 0:
    ga = (4 * tong_con - tong_chan) // 2
    cho = tong_con - ga
    if ga >= 0 and cho >= 0:
        print(f"Số lượng gà: {ga}, Số lượng chó: {cho}")
    else:
        print("invalid")
else:
    print("invalid")
```
