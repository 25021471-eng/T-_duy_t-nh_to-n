W8A16 Yêu cầu
Xây dựng lớp Shape với phương thức area().
Xây dựng lớp Rectangle(w, h) và Circle(r) kế thừa từ Shape, ghi đè area().
Lấy π = 3.14.
Input
Dòng 1: loại hình (RECTANGLE hoặc CIRCLE)
Dòng 2: tham số tương ứng.
Output
In ra diện tích với 2 chữ số thập phân. 


```python
import sys

class Shape:
    def area(self):
        pass

class Rectangle(Shape):
    def __init__(self, w, h):
        self.w = w
        self.h = h
        
    def area(self):
        return self.w * self.h

class Circle(Shape):
    def __init__(self, r):
        self.r = r
        
    def area(self):
        return 3.14 * (self.r ** 2)

def solve_w8a16():
    lines = sys.stdin.read().splitlines()
    if not lines:
        return
    shape_type = lines[0].strip()
    params = list(map(float, lines[1].split()))
    
    if shape_type == "RECTANGLE":
        shape = Rectangle(params[0], params[1])
    elif shape_type == "CIRCLE":
        shape = Circle(params[0])
        
    print(f"{shape.area():.2f}")
```

Xây dựng lớp Stack sử dụng danh sách để cài đặt ngăn xếp với các phương thức push(x),

Input
Dòng 1: n

PUSH x
POP
TOP
Output
Mỗi lệnh TOP in ra giá trị ở đỉnh ngăn xếp trên một dòng.

```python
import sys

class Stack:
    def __init__(self):
        self.items = []
        
    def push(self, x): 
        self.items.append(x)
        
    def pop(self): 
        if self.items:
            return self.items.pop()
            
    def top(self): 
        if self.items:
            return self.items[-1]

def solve_w8a30():
    lines = sys.stdin.read().splitlines()
    if not lines:
        return
    stack = Stack()
    for line in lines[1:]: 
        cmd = line.strip().split()
        if not cmd:
            continue
        
        if cmd[0] == "PUSH": 
            stack.push(int(cmd[1]))
        elif cmd[0] == "POP":
            stack.pop()
        elif cmd[0] == "TOP":
            print(stack.top()) 
```

W7A16
Cho một mảng số nguyên không âm. Gọi
mảng đó là "mảng đẹp" nếu mỗi phần tử
x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp
hay không. Nếu không hãy xóa đi y số (y
là nhỏ nhất có thể) để mảng còn lại thành
mảng đẹp

```python
from collections import Counter

def solve_w7a16(arr):
    counts = Counter(arr)
    deleted_count = 0
    
    for x, count in counts.items():
        if x == 0:
            deleted_count += count
        elif count > x:
            deleted_count += (count - x)
        elif count < x:
            deleted_count += count
            
    if deleted_count == 0:
        print("YES")
    else:
        print(f"NO {deleted_count}")
```

W7A19
Merge 2 mảng đã sắp xếp (có thể tăng
dần, có thể giảm dần) mà không dùng
bất cứ thuật toán sắp xếp nào thành một
mảng cũng có cùng thứ tự sắp xếp như 2
mảng ban đầu


```python
def solve_w7a19(arr1, arr2):
    def is_ascending(arr):
        for i in range(len(arr) - 1):
            if arr[i] < arr[i+1]: return True
            if arr[i] > arr[i+1]: return False
        return True

    asc = is_ascending(arr1) if len(arr1) > 1 else (is_ascending(arr2) if len(arr2) > 1 else True)
    
    i, j = 0, 0
    result = []
    
    while i < len(arr1) and j < len(arr2):
        if asc:
            if arr1[i] <= arr2[j]:
                result.append(arr1[i])
                i += 1
            else:
                result.append(arr2[j])
                j += 1
        else:
            if arr1[i] >= arr2[j]:
                result.append(arr1[i])
                i += 1
            else:
                result.append(arr2[j])
                j += 1
                
    result.extend(arr1[i:])
    result.extend(arr2[j:])
    print(*(result))
```

```python
def solve_w7a24(text):
    words = text.split()
    if not words:
        return ""
    
    result = [words[0]]
    for i in range(1, len(words)):
        clean_curr = ''.join(filter(str.isalnum, words[i])).lower()
        clean_prev = ''.join(filter(str.isalnum, result[-1])).lower()
        
        if clean_curr != clean_prev:
            result.append(words[i])
            
    print(" ".join(result))
```

W7A25

đổi vị trí các chữ số


```python
def solve_w7a25(num_str, k):
    stack = []
    to_remove = len(num_str) - k
    
    for digit in num_str:
        while to_remove > 0 and stack and stack[-1] < digit:
            stack.pop()
            to_remove -= 1
        stack.append(digit)
    result = "".join(stack[:k])
    print(result)
```

không

```python
def solve_w7a27(s):
    stack = []
    mapping = {')': '(', '}': '{', ']': '[', '>': '<'}
    
    for char in s:
        if char in mapping.values():
            stack.append(char)
        elif char in mapping.keys():
            if not stack or stack[-1] != mapping[char]:
                print("NO")
                return
            stack.pop()
            
    print("YES" if not stack else "NO")
```

nhất (gặp đầu tiên từ trái qua)


```python
def is_prime(n):
    if n < 2: return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0: return False
    return True

def solve_w7a28(n):
    max_len = 0
    best_seq = []
    current_seq = []
    
    for i in range(2, n + 1):
        if not is_prime(i): 
            current_seq.append(i)
        else:
            if len(current_seq) > max_len:
                max_len = len(current_seq)
                best_seq = current_seq
            current_seq = []
            
    if len(current_seq) > max_len:
        best_seq = current_seq
        
    print(", ".join(map(str, best_seq)))
```

```python
def solve_w7a29(arr):
    if not arr: return
    max_len = 1
    best_seq = [arr[0]]
    curr_seq = [arr[0]]
    
    for i in range(1, len(arr)):
        if arr[i] > arr[i-1]:
            curr_seq.append(arr[i])
        else:
            if len(curr_seq) > max_len:
                max_len = len(curr_seq)
                best_seq = curr_seq
            curr_seq = [arr[i]]
            
    if len(curr_seq) > max_len:
        best_seq = curr_seq
        
    print(*(best_seq))
```

W7A30
Tìm Dãy con tăng dần KHÔNG liên tục
dài nhất của một danh sách nhập vào từ
bàn phím 


```python
def solve_w7a30(arr):
    if not arr: return
    n = len(arr)
    dp = [1] * n
    parent = [-1] * n
    
    for i in range(1, n):
        for j in range(i):
            if arr[i] > arr[j] and dp[i] < dp[j] + 1:
                dp[i] = dp[j] + 1
                parent[i] = j
                
    max_len = max(dp)
    best_idx = dp.index(max_len)
    result = []
    while best_idx != -1:
        result.append(arr[best_idx])
        best_idx = parent[best_idx]
        
    result.reverse()
    print(*(result))
```

W7A31
Đổi chỗ đúng 2 chữ số để được số nhỏ
nhất có thể

```python
def solve_w7a31(num_str):
    num = list(num_str)
    n = len(num)
    for i in range(n - 1):
        min_idx = i
        for j in range(n - 1, i, -1):
            if i == 0 and num[j] == '0':
                continue
            if num[j] < num[min_idx]:
                min_idx = j
        if min_idx != i:
            num[i], num[min_idx] = num[min_idx], num[i]
            break
            
    print("".join(num))
```
