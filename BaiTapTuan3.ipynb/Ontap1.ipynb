Yêu cầu. Cước phí Taxi Thông Minh.
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả,
kiểu số nguyên, đơn vị VND).
Các quy tắc cần cài đặt:
1. Giá mở cửa (0–0.8 km):
- Xe 4 chỗ: 11,000 VND
- Xe 7 chỗ: 12,000 VND
2. Giá theo km sau 0.8 km:
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND.
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường +
chờ).
5. Điều kiện đặc biệt:
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu). 


```python
import math
import re
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    if car_type == 4:
        if km <= 0.8:
            fare = 11000
        elif km <= 20:
            fare = 11000 + (km - 0.8) * 13000
        else:
            fare = 11000 + (20 - 0.8) * 13000 + (km - 20) * 10500
    else: 
        if km <= 0.8:
            fare = 12000
        elif km <= 20:
            fare = 12000 + (km - 0.8) * 14000
        else:
            fare = 12000 + (20 - 0.8) * 14000 + (km - 20) * 12000
    wait_fee = 0
    if wait_minutes > 5:
        wait_fee = math.ceil((wait_minutes - 5) / 5) * 2000
        
    temp_total = fare + wait_fee
    is_night = start_hour in [22, 23, 0, 1, 2, 3, 4, 5]
    if is_night:
        temp_total *= 1.2
    if km > 30 and not is_night:
        temp_total *= 0.9
        
    if temp_total < 50000 and km > 5:
        temp_total = 50000
        
    return int(temp_total)
```

```python
def is_triangular_number(x):
    total = 0
    n = 1
    while total < x:
        total += n
        if total == x:
            return True
        n += 1
    return False
```

Yêu cầu: Cho các hệ số thực a, b, c, trả về tất cả các nghiệm thực của phương trình bậc hai axx + bx + c = 0 trong một danh
sách được sắp xếp tăng dần
Quy tắc:
- Nếu a == 0 và b != 0: trả về nghiệm duy nhất [-c / b].
- Nếu a == 0 và b == 0: trả về [] (không có nghiệm hữu hạn duy nhất nếu c == 0, hoặc vô nghiệm nếu c != 0).
- Với a != 0, dùng biệt thức d = bb - 4ac:
- d < 0 → []
- d == 0 → [ -b / (2a) ]
- d > 0 → hai nghiệm [r1, r2] với r1 <= r2.


```python
def solve_quadratic(a, b, c):
    if a == 0 and b != 0: return [-c/b]
    if a == 0 and b == 0: return []
    d = b*b - 4*a*c
    if d < 0: return []
    if d == 0: return [-b/(2*a)]
    r1, r2 = (-b - math.sqrt(d))/(2*a), (-b + math.sqrt(d))/(2*a)
    return sorted([r1, r2])
```

Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3). 


```python
def angle_abc(x1, y1, x2, y2, x3, y3):
    vba = (x1 - x2, y1 - y2)
    vbc = (x3 - x2, y3 - y2)
    dot_product = vba[0]*vbc[0] + vba[1]*vbc[1]
    mag_ba = math.hypot(*vba)
    mag_bc = math.hypot(*vbc)
    cos_angle = dot_product / (mag_ba * mag_bc)
    return math.degrees(math.acos(round(cos_angle, 10)))
```

Yêu cầu: Viết hàm are_collinear(x1, y1, x2, y2, x3, y3) trả về True nếu ba điểm thẳng hàng, ngược lại trả về False.
- Đầu vào: ba điểm x1, y1, x2, y2, x3, y3 biểu diễn tọa độ của ba điểm.
- Đầu ra: giá trị boolean. 
- Gợi ý: Dùng phép kiểm tra diện tích (tích có hướng) và một sai số nhỏ khi so sánh số thực. Các điểm nằm trên cùng một
đường thẳng được xem là thẳng hàng (kể cả khi có điểm trùng nhau). 

```python
def point_in_quadrilateral(quad, p): # Bài 25
    x, y = p
    inside = False
    n = len(quad)
    for i in range(n):
        x1, y1 = quad[i]
        x2, y2 = quad[(i + 1) % n]
        if min(x1, x2) <= x <= max(x1, x2) and min(y1, y2) <= y <= max(y1, y2):
            if abs((x2 - x1) * (y - y1) - (x - x1) * (y2 - y1)) < 1e-9:
                return True
        if ((y1 > y) != (y2 > y)) and (x < (x2 - x1) * (y - y1) / (y2 - y1) + x1):
            inside = not inside
    return inside
```

Yêu cầu: Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình ax^2 + bx + c = 0.
- Đầu vào: các số thực a, b, c.

- Quy tắc:

- Nếu a == 0 và b == 0 và c != 0, phương trình vô nghiệm. 

```python
def are_collinear(x1, y1, x2, y2, x3, y3): # Bài 26
    area = x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)
    return abs(area) < 1e-9
```

Yêu cầu: Viết hàm compute(a, b, op) tính a (+|-||/) b tùy theo op.
- Đầu vào: các số a, b và ký tự op thuộc {'+', '-', '', '/'}.
- Đầu ra: kết quả số học.
- Yêu cầu thêm:
- Nếu op là '/' và b == 0, phát sinh ValueError.
- Nếu op không hợp lệ, phát sinh ValueError.

```python
def is_right_triangle(a, b, c): # Bài 32
    if a <= 0 or b <= 0 or c <= 0: return False
    sides = sorted([a, b, c])
    # Kiểm tra bất đẳng thức tam giác và định lý Pythagore
    if sides[0] + sides[1] > sides[2]:
        return abs(sides[0]**2 + sides[1]**2 - sides[2]**2) < 1e-9
    return False
```

```python
def compute(a, b, op): # Bài 33
    if op == '+': return a + b
    if op == '-': return a - b
    if op == '*': return a * b
    if op == '/':
        if b == 0: raise ValueError("Division by zero")
        return a / b
    raise ValueError("Invalid operator")
```

```python
def to_lower_single(c): # Bài 40
    if len(c) != 1 or not c.isalpha():
        raise ValueError("Input must be a single character")
    return c.lower()
```

Viết hàm trả về True nếu a, b, c có thể là độ dài ba cạnh của một tam giác; ngược lại trả về False. 


```python
def shift_char(x, n): # Bài 41
    if not x.isalpha() or len(x) != 1: return x
    base = ord('A') if x.isupper() else ord('a')
    return chr((ord(x) - base + n) % 26 + base)
```

Viết hàm tính diện tích tam giác có ba cạnh a, b, c (dùng công thức Heron). Phát sinh ValueError nếu ba cạnh không thể
tạo thành tam giác.

```python
def is_valid_triangle(a, b, c): # Bài 42
    return a + b > c and a + c > b and b + c > a
```

Viết hàm trả về tích các số nguyên từ 1 đến n (giai thừa). Với n=0, trả về 1. 

```python
def heron_area(a, b, c): # Bài 43
    if not is_valid_triangle(a, b, c):
        raise ValueError("Not a triangle")
    p = (a + b + c) / 2
    return math.sqrt(p * (p - a) * (p - b) * (p - c))
```

Viết hàm trả về số lượng số nguyên dương k sao cho 1 <= k < n và k > (n - 1) / 2. (Ví dụ: n=9 -> 4). 


```python
def factorial(n): # Bài 44
    if n == 0: return 1
    return math.prod(range(1, n + 1))
```

```python
def count_k(n): # Bài 45
    count = 0
    for k in range(1, n):
        if k > (n - 1) / 2:
            count += 1
    return count
```
