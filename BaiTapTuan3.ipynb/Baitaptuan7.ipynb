Viết hàm tìm kiếm nhị phân
trên một danh sách đã
được sắp xếp tăng dần.
Trả về chỉ số của phần tử
hoặc -1 nếu không tìm
thấy. mẫu hàm: def
binary_search(arr, left,
right, target)

```python
def binary_search(arr, left, right, target):
    while left <= right:
        mid = left + (right - left) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

W7A3
Viết hàm thực hiện thuật
toán bubble sort, Selection
Sort để sắp xếp danh sách
theo thứ tự tăng dần. Sắp
xếp nổi bọt (Bubble Sort); 

```python
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```

```python
def most_frequent(arr):
    if not arr:
        return None
    
    counts = {}
    first_idx = {}
    
    for idx, val in enumerate(arr):
        counts[val] = counts.get(val, 0) + 1
        if val not in first_idx:
            first_idx[val] = idx
    best_item = max(counts.keys(), key=lambda x: (counts[x], -first_idx[x]))
    
    print(f"{best_item} xuat hien nhieu nhat, som nhat, {counts[best_item]} lan")
    return best_item
```

tất cả các cặp (a, b) sao

```python
def count_pairs_with_sum(arr, X):
    from collections import Counter
    freq = Counter(arr)
    count = 0
    
    for num in freq:
        complement = X - num
        if complement == num:
            count += freq[num] * (freq[num] - 1)
        elif complement > num: 
            if complement in freq:
                count += freq[num] * freq[complement]
                
    return count
```

W7A7
Cho một danh sách các số
nguyên nums, hãy tìm độ
dài của dãy con tăng
nghiêm ngặt dài nhất
(không nhất thiết phải liền kề).

```python
import bisect

def length_of_lis(nums):
    if not nums:
        return 0
    tails = []
    for num in nums:
        idx = bisect.bisect_left(tails, num)
        if idx == len(tails):
            tails.append(num)
        else:
            tails[idx] = num
    return len(tails)
```

W7A8
Bạn được cung cấp: một
danh sách các khoảng thời
gian interval = [[l1, r1], [l2,
r2], ...] và một danh sách
các truy vấn queries = [q1,
q2, ...]
Với mỗi truy vấn q, hãy tìm
độ dài nhỏ nhất của
khoảng thời gian [li, ri] sao
cho li ≤ q ≤ ri.
Nếu không tồn tại khoảng
thời gian nào như vậy, hãy
trả về -1 cho truy vấn đó.
solve in < O(m.n), hint:
binary search in the
queries list


```python
import heapq

def min_interval(intervals, queries):
    intervals.sort(key=lambda x: x[0])
    sorted_queries = sorted(enumerate(queries), key=lambda x: x[1])
    res = [-1] * len(queries)
    
    heap = [] 
    i = 0
    n = len(intervals)
    
    for original_idx, q in sorted_queries:
        while i < n and intervals[i][0] <= q:
            l, r = intervals[i]
            heapq.heappush(heap, (r - l + 1, r))
            i += 1
        while heap and heap[0][1] < q:
            heapq.heappop(heap)
        if heap:
            res[original_idx] = heap[0][0]
            
    return res
```

W7A9
Cho một danh sách các số
nums, với mỗi phần tử
nums[i], hãy đếm xem có
bao nhiêu phần tử bên phải
nhỏ hơn nums[i].
Trả về danh sách các số
đếm.
Efficient solution: using
heap or binary search tree


```python
class FenwickTree:
    def __init__(self, size):
        self.tree = [0] * (size + 1)
        
    def update(self, i, delta):
        while i < len(self.tree):
            self.tree[i] += delta
            i += i & (-i)
            
    def query(self, i):
        s = 0
        while i > 0:
            s += self.tree[i]
            i -= i & (-i)
        return s

def count_smaller(nums):
    ranks = {val: i + 1 for i, val in enumerate(sorted(set(nums)))}
    
    ft = FenwickTree(len(ranks))
    res = []
    for num in reversed(nums):
        rank = ranks[num]
        res.append(ft.query(rank - 1))
        ft.update(rank, 1)
        
    return res[::-1]
```

W7A10
Cho một danh sách các số
nguyên nums và một số
nguyên m, hãy tìm tổng
mảng con lớn nhất theo
modulo m.
Tức là, với bất kỳ mảng
con nums[i:j] nào, hãy tính
sum(nums[i:j]) % m và trả
về kết quả lớn nhất có thể.
Gợi ý:
Duy trì tổng tiền tố (prefix
sum) mod m và sử dụng
tìm kiếm nhị phân để tìm
tổng tiền tố nhỏ nhất chỉ
lớn hơn tổng hiện tại một
chút (để tối đa hóa hiệu số
bao quanh).

```python
import bisect

def max_subarray_mod(nums, m):
    prefix_set = [0]
    curr_sum = 0
    max_mod = 0
    
    for num in nums:
        curr_sum = (curr_sum + num) % m
        max_mod = max(max_mod, curr_sum)
        idx = bisect.bisect_right(prefix_set, curr_sum)
        if idx < len(prefix_set):
            max_mod = max(max_mod, (curr_sum - prefix_set[idx] + m) % m)
            
        bisect.insort(prefix_set, curr_sum)
        
    return max_mod
```

Input: 1 3 2 2 5 2 3 7
Output: 5 - Dãy hài hòa con
dài nhất là [3,2,2,2,3]


```python
from collections import Counter

def find_lhs(arr):
    count = Counter(arr)
    max_length = 0
    for key in count:
        if key + 1 in count:
            max_length = max(max_length, count[key] + count[key + 1])
    return max_length
```

Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu
mỗi phần tử x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp.


```python
from collections import Counter

def make_beautiful_array(arr):
    counts = Counter(arr)
    min_deletions = 0
    
    for x, count in counts.items():
        if count >= x:
            min_deletions += (count - x)
        else:
            min_deletions += count
            
    return min_deletions
```

W7A19
Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không
dùng bất cứ thuật toán sắp xếp nào


```python
def merge_sorted_arrays(arr1, arr2):
    if len(arr1) > 1 and arr1[0] > arr1[-1]: arr1 = arr1[::-1]
    if len(arr2) > 1 and arr2[0] > arr2[-1]: arr2 = arr2[::-1]
    
    res = []
    i, j = 0, 0
    
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j]:
            res.append(arr1[i])
            i += 1
        else:
            res.append(arr2[j])
            j += 1
    res.extend(arr1[i:])
    res.extend(arr2[j:])
    return res
```

W7A24 Cho xâu kí tự, xóa các từ giống nhau liền kề, chỉ giữ lại từ đầu tiên


```python
def remove_adjacent_duplicates(text):
    words = text.split()
    if not words: return ""
    
    res = [words[0]]
    for i in range(1, len(words)):
        if words[i] != words[i-1]:
            res.append(words[i])
            
    return " ".join(res)
```

W7A25
Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi
đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số


```python
def max_number_from_string(num_str, k):
    n = len(num_str)
    to_remove = n - k  
    stack = []
    
    for digit in num_str:
        while to_remove > 0 and stack and stack[-1] < digit:
            stack.pop()
            to_remove -= 1
        stack.append(digit)
    return "".join(stack[:k])
```

Cho một dãy ngoặc ()(){}{}[][]..., kiểm tra xem dãy ngoặc đã cho có hợp lệ
hay không

```python
def is_valid_brackets(s):
    stack = []
    mapping = {")": "(", "}": "{", "]": "["}
    
    for char in s:
        if char in mapping:
            top_element = stack.pop() if stack else '#'
            if mapping[char] != top_element:
                return False
        else:
            stack.append(char)
            
    return len(stack) == 0
```

̀i toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên
nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ
dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua)


```python
def longest_composite_sequence(n):
    is_prime = [True] * (n + 1)
    if n >= 0: is_prime[0] = False
    if n >= 1: is_prime[1] = False
    
    for i in range(2, int(n**0.5) + 1):
        if is_prime[i]:
            for j in range(i*i, n + 1, i):
                is_prime[j] = False
                
    max_len = 0
    best_start = -1
    curr_len = 0
    curr_start = -1
    
    for i in range(2, n + 1):
        if not is_prime[i]: 
            if curr_len == 0:
                curr_start = i
            curr_len += 1
        else:
            if curr_len > max_len:
                max_len = curr_len
                best_start = curr_start
            curr_len = 0
            
    if curr_len > max_len:
        max_len = curr_len
        best_start = curr_start
        
    if max_len == 0: return []
    return list(range(best_start, best_start + max_len))
```

Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau


```python
def longest_continuous_increasing_subsequence(arr):
    if not arr: return []
    
    max_len = 1
    best_start = 0
    curr_len = 1
    curr_start = 0
    
    for i in range(1, len(arr)):
        if arr[i] > arr[i-1]:
            curr_len += 1
        else:
            if curr_len > max_len:
                max_len = curr_len
                best_start = curr_start
            curr_len = 1
            curr_start = i
            
    if curr_len > max_len:
        max_len = curr_len
        best_start = curr_start
        
    return arr[best_start : best_start + max_len]
```

Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập
vào từ bàn phím 

```python
def longest_increasing_subsequence_v2(arr):
    if not arr: return []
    n = len(arr)
    dp = [1] * n
    parent = [-1] * n 
    
    for i in range(1, n):
        for j in range(0, i):
            if arr[i] > arr[j] and dp[i] < dp[j] + 1:
                dp[i] = dp[j] + 1
                parent[i] = j
                
    max_len = max(dp)
    best_idx = dp.index(max_len)
    result = []
    while best_idx != -1:
        result.append(arr[best_idx])
        best_idx = parent[best_idx]
        
    return result[::-1]
```
