W9A1 Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ thuật toán sắp xếp nào 


```python
def merge_sorted_lists(list1, list2):
    i, j = 0, 0
    result = []
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            result.append(list1[i])
            i += 1
        else:
            result.append(list2[j])
            j += 1
    result.extend(list1[i:])
    result.extend(list2[j:])
    return result
```

W9A2
Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên dòng tăng dần và trên cột
tăng dần, hãy tìm số lớn thứ k (k < nxm) với thuật toán tốt nhất có thể

```python
import heapq

def find_kth_largest(matrix, k):
    n = len(matrix)
    m = len(matrix[0])
    min_heap = []
    for r in range(n):
        for c in range(m):
            heapq.heappush(min_heap, matrix[r][c])
            if len(min_heap) > k:
                heapq.heappop(min_heap)
    return min_heap[0]
```

Viết hàm canPlaceFlowers(flowerbed, k) kiểm tra xem có thể trồng hết số hoa mới được
cấp vào trong luống đất của bạn hay không.
Hàm nhận đầu vào là một danh sách (list) flowerbed, đánh dấu những vị trí trong luống đất
đã được trồng cây xanh trước đó (giá trị 1 ứng với vị trí cây đã được trồng và 0 ứng với vị
trí đất trống) và số lượng hoa được phát thêm k (int). Hàm trả về True nếu có thể trồng hết
toàn bộ k cây hoa vào luống đất và False trong trường hợp ngược lại. 


```python
def canPlaceFlowers(flowerbed, k):
    count = 0
    n = len(flowerbed)
    for i in range(n):
        if flowerbed[i] == 0:
            left_empty = (i == 0) or (flowerbed[i-1] == 0)
            right_empty = (i == n - 1) or (flowerbed[i+1] == 0)
            if left_empty and right_empty:
                flowerbed[i] = 1
                count += 1
                if count >= k:
                    return True
    return count >= k
```

Ban đầu, một con rô-bốt ở vị trí (0, 0) trên một lưới 2D.
Cho một chuỗi `moves` chứa các thao tác di chuyển của rô-bốt.
Các bước di chuyển được biểu diễn bởi các ký tự:
- 'R' (Phải)
- 'L' (Trái)
- 'U' (Lên)
- 'D' (Xuống)
Viết hàm `def is_path_crossing(moves)`:
Hàm nhận đầu vào là một chuỗi `moves`.
Hàm trả về `True` nếu trong quá trình di chuyển, rô-bốt đi vào một ô mà nó đã từng đi qua
trước đó.
Hàm trả về `False` nếu rô-bốt không bao giờ đi lặp lại (luôn đi vào ô mới).
Lưu ý: Vị trí ban đầu (0, 0) được tính là vị trí đầu tiên đã đi qua. 


```python
def is_path_crossing(moves):
    visited = {(0, 0)}
    x, y = 0, 0
    for move in moves:
        if move == 'U':
            y += 1
        elif move == 'D':
            y -= 1
        elif move == 'R':
            x += 1
        elif move == 'L':
            x -= 1
        if (x, y) in visited:
            return True
        visited.add((x, y))
    return False
```

[Tìm bán kính lò sưởi nhỏ nhất]
Mùa đông đang đến! Bạn được cung cấp vị trí của các ngôi nhà và các lò sưởi trên một
đường thẳng.
Nhiệm vụ của bạn là tìm bán kính r nhỏ nhất cho các lò sưởi sao cho mọi ngôi nhà đều
nằm trong phạm vi sưởi ấm của ít nhất một lò sưởi.
Tất cả các lò sưởi đều có cùng một bán kính r.
Một ngôi nhà tại vị trí h được sưởi ấm bởi một lò sưởi tại vị trí t nếu |h - t| <= r.
Viết hàm `def findRadius(houses, heaters):`
Hàm nhận đầu vào là hai danh sách (list) các số nguyên:
- `houses`: chứa vị trí của các ngôi nhà.
- `heaters`: chứa vị trí của các lò sưởi.
Hàm trả về một số nguyên là bán kính r nhỏ nhất cần thiết.
Lưu ý:
- Số lượng nhà và lò sưởi là một số không âm và không vượt quá 25000.
- Vị trí của các ngôi nhà và lò sưởi là một số không âm và không vượt quá 10^9. 


```python
import sys

def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()
    r = 0
    i = 0
    m = len(heaters)
    for house in houses:
        while i < m - 1 and abs(heaters[i+1] - house) <= abs(heaters[i] - house):
            i += 1
        r = max(r, abs(heaters[i] - house))
    return r
```

Yêu cầu:
- Không dùng vòng lặp
- Không dùng hàm có sẵn

```python
def power(base, exp):
    if exp == 0:
        return 1
    half = power(base, exp // 2)
    if exp % 2 == 0:
        return half * half
    else:
        return base * half * half
```

```python
def clean_and_title_case(s):
    result = []
    word = []
    for char in s:
        if char != ' ' and char != '\t' and char != '\n':
            word.append(char)
        else:
            if word:
                result.append("".join(word))
                word = []
    if word:
        result.append("".join(word))
        
    final_words = []
    for w in result:
        capitalized_word = ""
        first_char = w[0]
        if 'a' <= first_char <= 'z':
            capitalized_word += chr(ord(first_char) - 32)
        else:
            capitalized_word += first_char
            
        for i in range(1, len(w)):
            char = w[i]
            if 'A' <= char <= 'Z':
                capitalized_word += chr(ord(char) + 32)
            else:
                capitalized_word += char
        final_words.append(capitalized_word)
        
    final_str = ""
    for i in range(len(final_words)):
        final_str += final_words[i]
        if i < len(final_words) - 1:
            final_str += " "
    return final_str
```

W9A21
Tìm phần tử lớn thứ k trong dãy n phần tử bằng thuật toán tốt nhất có thể – Tránh dùng sắp xếp
(dùng phân hoạch của quicksort) 

```python
def partition(arr, low, high):
    pivot = arr[high]
    i = low
    for j in range(low, high):
        if arr[j] >= pivot:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1
    arr[i], arr[high] = arr[high], arr[i]
    return i

def quick_select(arr, low, high, k):
    if low <= high:
        pivot_index = partition(arr, low, high)
        if pivot_index == k - 1:
            return arr[pivot_index]
        elif pivot_index > k - 1:
            return quick_select(arr, low, pivot_index - 1, k)
        else:
            return quick_select(arr, pivot_index + 1, high, k)
    return -1

def find_kth_largest_quickselect(arr, k):
    return quick_select(arr, 0, len(arr) - 1, k)
```

Hai chuỗi là anagram nếu chúng có cùng ký tự và số lần xuất hiện. viết CT kiểm tra xem 2 chuỗi
đầu vào có phải là anagram của nhau không 


```python
def check_anagram(str1, str2):
    if len(str1) != len(str2):
        return False
    
    counts = {}
    for char in str1:
        counts[char] = counts.get(char, 0) + 1
        
    for char in str2:
        if char not in counts:
            return False
        counts[char] -= 1
        if counts[char] < 0:
            return False
            
    return True
```

W9A23
Cho số nguyên rất lớn (string), và một số nguyên dương k. Tìm số có k chữ số lớn nhất có thể
từ số ban đầu bằng cách xóa bớt đi một số chữ số nào đó mà không thay đổi vị trí của bất kì
chữ số nào. (VD: 78934340999, k = 7 => 9440999 là số lớn nhất có 7 chữ số cần tìm) 


```python
def largest_number_deleting_digits(num_str, k):
    stack = []
    to_remove = len(num_str) - k
    
    for digit in num_str:
        while to_remove > 0 and stack and stack[-1] < digit:
            stack.pop()
            to_remove -= 1
        stack.append(digit)
        
    final_digits = stack[:k]
    return "".join(final_digits)
```

W9A24
Nén chuối kí tự bằng cách đếm số lần xuất hiện của các kí tự liên tiếp nhau và ghi lại ngay sau
kí tự đó. VD aaaaaafdhhhhh =>a6f1d1h5 

```python
def compress_string(s):
    if not s:
        return ""
        
    result = []
    current_char = s[0]
    count = 1
    
    for i in range(1, len(s)):
        if s[i] == current_char:
            count += 1
        else:
            result.append(current_char + str(count))
            current_char = s[i]
            count = 1
            
    result.append(current_char + str(count))
    return "".join(result)
```

Giả sử mật khẩu mạnh là một chuỗi kí tự dài ít nhất 8 kí tự gồm ít nhất 1 chữ cái thường, 1 chữ
cái hoa, 1 chữ số và 1 trong các kí tự đặc biệt !, @, #, $. Nhập vào một chuỗi kĩ tự, kiểm tra xem
nó có phải là mật khẩu mạnh không. 

```python
def is_strong_password(password):
    if len(password) < 8:
        return False
        
    has_lower = False
    has_upper = False
    has_digit = False
    has_special = False
    special_chars = {'!', '@', '#', '$'}
    
    for char in password:
        if 'a' <= char <= 'z':
            has_lower = True
        elif 'A' <= char <= 'Z':
            has_upper = True
        elif '0' <= char <= '9':
            has_digit = True
        elif char in special_chars:
            has_special = True
            
    return has_lower and has_upper and has_digit and has_special
```
