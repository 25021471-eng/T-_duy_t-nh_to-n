Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_armstrong(x: int) -> bool:
    if x < 0:
        return False
    s = str(x)
    k = len(s)
    total = sum(int(digit) ** k for digit in s)
    return total == x
```

Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
def is_perfect_square(x: int) -> bool:
    if x < 0:
        return False
    root = int(x ** 0.5)
    return root * root == x
```

W5A18 VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên.


```python
def is_palindrome(x: int) -> bool:
    if x < 0:
        return False
    return str(x) == str(x)[::-1]
```

W5A19 VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên

```python
def is_magic_number(x: int) -> bool:
    if x <= 0:
        return False
    while x >= 10:
        x = sum(int(d) for d in str(x))
    return x == 1
```

W5A20 Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.


```python
def detect_number_type(s: str) -> str:
    try:
        int(s)
        return 'int'
    except ValueError:
        pass
    try:
        float(s)
        return 'float'
    except ValueError:
        pass
        
    return 'neither'
```

W5A21 Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
def is_one_product_of_others(a, b, c) -> bool:
    return (a == b * c) or (b == a * c) or (c == a * b)
```

W5A22 Viết hàm sum_halves(s) nhận chuỗi 6 chữ số và trả về tổng của hai số 3 chữ số đầu và 3 chữ số cuối.


```python
def sum_halves(s: str) -> int:
    if len(s) != 6 or not s.isdigit():
        raise ValueError("Chuỗi phải có đúng 6 chữ số.")
    first_half = int(s[:3])
    second_half = int(s[3:])
    return first_half + second_half
```

W5A23 Viết hàm quadratic_roots_count(a, b, c) trả về số nghiệm thực của phương trình a*x^2 + b*x + c = 0.


```python
def quadratic_roots_count(a, b, c) -> int:
    if a == 0:
        if b == 0:
            return -1 if c == 0 else 0  
        return 1
    
    delta = b**2 - 4*a*c
    if delta > 0:
        return 2
    elif delta == 0:
        return 1
    else:
        return 0
```

W5A24 Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không

```python
def are_collinear(x1, y1, x2, y2, x3, y3) -> bool:
    return (x1 * (y2 - y3) + x2 * (y3 - y1) + x3 * (y1 - y2)) == 0
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)

```python
def point_in_quadrilateral(quad: list, p: tuple) -> bool:
    x, y = p
    inside = False
    n = len(quad)
    p1x, p1y = quad[0]
    for i in range(n + 1):
        p2x, p2y = quad[i % n]
        if y > min(p1y, p2y):
            if y <= max(p1y, p2y):
                if x <= max(p1x, p2x):
                    if p1y != p2y:
                        xints = (y - p1y) * (p2x - p1x) / (p2y - p1y) + p1x
                    if p1x == p2x or x <= xints:
                        inside = not inside
        p1x, p1y = p2x, p2y
    return inside
```

W5A27 Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def count_non_zero_digits(numbers: list) -> int:
    count = 0
    for num in numbers:
        s = str(abs(num))
        count += sum(1 for d in s if d != '0')
    return count
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).

```python
import math

def calculate_angle_ABC(x1, y1, x2, y2, x3, y3) -> float:
    ba_x, ba_y = x1 - x2, y1 - y2
    bc_x, bc_y = x3 - x2, y3 - y2
    dot_product = ba_x * bc_x + ba_y * bc_y
    mag_ba = math.sqrt(ba_x**2 + ba_y**2)
    mag_bc = math.sqrt(bc_x**2 + bc_y**2)
    
    if mag_ba == 0 or mag_bc == 0:
        return 0.0 
    cos_angle = dot_product / (mag_ba * mag_bc)
    cos_angle = max(-1.0, min(1.0, cos_angle))
    
    angle_rad = math.acos(cos_angle)
    return math.degrees(angle_rad)
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k).


```python
import math

def binomial_coefficient(n: int, k: int) -> int:
    if k < 0 or k > n:
        return 0
    return math.comb(n, k)
```

Viết hàm smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính tổng tiền taxi theo quy tắc giá mở cửa,
giá km, phí chờ, phụ phí ban đêm và giảm giá.

```python
def smart_taxi_fare(car_type: str, start_hour: int, km: float, wait_minutes: int) -> float:
  
    base_fare = 12000    
    price_per_km = 15000 
    wait_fee_per_min = 500 
    total = base_fare + (km * price_per_km) + (wait_minutes * wait_fee_per_min)
    if start_hour >= 22 or start_hour < 5:
        total *= 1.10 
    if km > 30:
        total *= 0.95
        
    return total
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0.

```python
def sum_1_to_n(n: int) -> int:
    if n <= 0:
        return 0
        
    total_sum = 0
    for x in range(1, n + 1):
        s = str(x)
        if len(s) == len(set(s)):
            total_sum += x
            
    return total_sum
```
