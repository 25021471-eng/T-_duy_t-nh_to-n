Bài 2: Nhập vào bán kính của hình tròn. In ra màn hình diện tích và chu vi của nó.

```python
import math

r = float(input("Nhập bán kính hình tròn: "))

if r <= 0:
    print("Bán kính phải lớn hơn 0!")
else:
    chu_vi = 2 * math.pi * r
    dien_tich = math.pi * (r ** 2)
    print(f"Chu vi: {chu_vi}")
    print(f"Diện tích: {dien_tich}")
```

Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
import math

a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))
if (a + b > c) and (a + c > b) and (b + c > a) and a > 0 and b > 0 and c > 0:
    if a == b == c:
        print("Đây là Tam giác đều")
    elif a == b or b == c or a == c:
        if math.isclose(a**2 + b**2, c**2) or math.isclose(a**2 + c**2, b**2) or math.isclose(b**2 + c**2, a**2):
            print("Đây là Tam giác vuông cân")
        else:
            print("Đây là Tam giác cân")
    elif math.isclose(a**2 + b**2, c**2) or math.isclose(a**2 + c**2, b**2) or math.isclose(b**2 + c**2, a**2):
        print("Đây là Tam giác vuông")
    else:
        print("Đây là Tam giác thường")
    chu_vi = a + b + c
    p = chu_vi / 2
    dien_tich = math.sqrt(p * (p - a) * (p - b) * (p - c))
    
    print(f"Chu vi: {chu_vi}")
    print(f"Diện tích: {dien_tich}")
else:
    print("Khong la ba canh cua mot tam giac")
```

Bài 8: Giải hệ phương trình tuyến tính:
ax  by  m
cx  dy  n

```python
a = float(input("Nhập a: "))
b = float(input("Nhập b: "))
m = float(input("Nhập m: "))
c = float(input("Nhập c: "))
d = float(input("Nhập d: "))
n = float(input("Nhập n: "))

D = a * d - b * c
Dx = m * d - b * n
Dy = a * n - m * c

if D != 0:
    x = Dx / D
    y = Dy / D
    print(f"Hệ phương trình có nghiệm duy nhất: x = {x}, y = {y}")
else:
    if Dx == 0 and Dy == 0:
        print("Hệ phương trình có vô số nghiệm")
    else:
        print("Hệ phương trình vô nghiệm")
```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây. 

```python
x = int(input("Nhập số giây x: "))

if x < 0:
    print("Thời gian không được âm!")
else:
    gio = x // 3600
    phut = (x % 3600) // 60
    giay = x % 60
    
    print(f"{x} giây bằng {gio} giờ, {phut} phút, {giay} giây.")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
import math

xc = float(input("Nhập hoành độ tâm đường tròn: "))
yc = float(input("Nhập tung độ tâm đường tròn: "))
r = float(input("Nhập bán kính đường tròn: "))

xa = float(input("Nhập hoành độ điểm A: "))
ya = float(input("Nhập tung độ điểm A: "))
khoang_cach_binh_phuong = (xa - xc)**2 + (ya - yc)**2
r_binh_phuong = r**2

if math.isclose(khoang_cach_binh_phuong, r_binh_phuong):
    print("Điểm A nằm trên đường tròn.")
elif khoang_cach_binh_phuong < r_binh_phuong:
    print("Điểm A nằm trong đường tròn.")
else:
    print("Điểm A nằm ngoài đường tròn.")
```

```python
x = float(input("Nhập cơ số x: "))
y = float(input("Nhập số mũ y: "))

try:
    ket_qua = x ** y
    print(f"Kết quả {x}^{y} = {ket_qua}")
except ValueError:
    print("Không xác định (Ví dụ: tính căn bậc chẵn của một số âm).")
except OverflowError:
    print("Kết quả quá lớn, vượt giới hạn tính toán của máy tính.")
```
